```python
# 이진화
img = cv2.imread('./data/Lenna.png', cv2.IMREAD_GRAYSCALE)
cv2.imshow('Original', img)

# The method returns two outputs.
# The first is the threshold that was used and the second output is the thresholded image.
# 임계값 직접 설정(임계값: 100), 해당 픽셀이 임계값보다 크면 최대값 255로 설정
ret, img_binary = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)
# 최적의 임계값을 찾아주는 otsu 알고리즘 사용
ret, img_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

cv2.imshow('binary', img_binary)
cv2.imshow('otsu', img_otsu)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 적응적 이진화
img = cv2.imread('./data/sudoku.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imshow('Original', img)

block_size = 9
C = 5
ret, img_otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
img_ab1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, C)
img_ab2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, C)

print(ret)
cv2.imshow('binary', img_otsu)
cv2.imshow('ADAPTIVE_THRESH_MEAN_C', img_ab1)
cv2.imshow('ADAPTIVE_THRESH_GAUSSIAN_C', img_ab2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 그레이스케일 영상의 히스토그램
img = cv2.imread('./data/Lenna.png', cv2.IMREAD_GRAYSCALE)
cv2.imshow('Original', img)

# 히스토그램 그리기
hist = cv2.calcHist(images=[img], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist)

print("histogram shape: ", hist.shape)
print("histogram sum: ", hist.sum())
print("image shape: ", img.shape)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 그레이스케일 영상의 히스토그램 평활화
img = cv2.imread('./data/hawkes.bmp', cv2.IMREAD_GRAYSCALE)
rows, cols = img.shape[:2]
print(img.shape)
# 히스토그램 평활화 계산
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
# 누적 히스토그램
cdf = hist.cumsum()
# 0 값을 NaN으로 제거
cdf_m = np.ma.masked_equal(cdf, 0)
# 평활화 히스토그램 계산
cdf_m = (cdf_m - cdf_m.min()) / (rows * cols) * 255
# NaN을 다시 0으로 변환
cdf = np.ma.filled(cdf_m, 0).astype('uint8')
# 히스토그램을 픽셀로 매핑
img2 = cdf[img]
# openCV API 적용
img3 = cv2.equalizeHist(img)

# 평활화 결과 히스토그램 계산
hist2 = cv2.calcHist([img2], [0], None, [256], [0, 256])
hist3 = cv2.calcHist([img3], [0], None, [256], [0, 256])

cv2.imshow('Before', img)
cv2.imshow('Manual', img2)
cv2.imshow('equalizeHist', img3)

hists = {'Before': hist, 'Manual': hist2, 'equalizeHist': hist3}
for num, (key, value) in enumerate(hists.items()):
    plt.subplot(1, 3, num+1)
    plt.title(key)
    plt.plot(value)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 컬러 영상의 히스토그램 평활화
img = cv2.imread('./data/hawkes.bmp', cv2.IMREAD_COLOR)

# 컬러공간을 HSV로 변경
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# V (밝기값) 공간에 히스토그램 평활화 적용
img_hsv[:, :, 2] = cv2.equalizeHist(img_hsv[:, :, 2])

# 컬러공간을 RGB로 다시 변경
img2 = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)

cv2.imshow('Before', img)
cv2.imshow('After', img2)

# 평활화 결과 히스토그램 계산
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([img2], [0], None, [256], [0, 256])

hists = {'Before': hist, 'After': hist2}
for num, (key, value) in enumerate(hists.items()):
    plt.subplot(1, 2, num+1)
    plt.title(key)
    plt.plot(value)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 그레이스케일 영상의 히스토그램 스트레칭
img = cv2.imread('./data/hawkes.bmp', cv2.IMREAD_GRAYSCALE)

# 히스토그램 스트레칭 계산
img_f = img.astype(np.float32)
img_norm = ((img_f - img_f.min()) / (img_f.max() - img_f.min())) * 255
img_norm = img_norm.astype(np.uint8)

# 스트레칭 적용
img_norm2 = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX)

# 히스토그램 계산
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_norm = cv2.calcHist([img_norm], [0], None, [256], [0, 256])
hist_norm2 = cv2.calcHist([img_norm2], [0], None, [256], [0, 256])

cv2.imshow('Before', img)
cv2.imshow('Manual', img_norm)
cv2.imshow('normalize', img_norm2)

hists = {'Before': hist, 'Manual': hist_norm, 'normalize': hist_norm2}
for num, (key, value) in enumerate(hists.items()):
    plt.subplot(1, 3, num+1)
    plt.title(key)
    plt.plot(value)
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 합성곱으로 블러링
img = cv2.imread('./data/Lenna.png', cv2.IMREAD_GRAYSCALE)
# 필터 마스크 행렬 정의
kernel = np.ones((3, 3), dtype = np.float64) / 9
# src: 입력영상, input image
# ddepth: 출력영상(원하는 depth 지정 가능, -1은 입력영상과 같은 형태), desired depth of the destination image
# kernel: 필터 마스크 행렬, 실수값, convolution kernel (or rather a correlation kernel)
# anchor: 고정점 위치, default value (-1,-1) means that the anchor is at the kernel center
# delta: 추가적으로 더할 값, optional value added to the filtered pixels
# borderType: 가장자리 픽셀 확장 방식, pixel extrapolation method
result = cv2.filter2D(src=img, ddepth=-1, kernel=kernel)

cv2.imshow('Original', img)
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
# 영상이 부드러워짐
```

```python
img = cv2.imread('./data/Lenna.png', cv2.IMREAD_GRAYSCALE)
# src: 입력영상
# ksize: 가우시안 필터 크기(가로, 세로), (0, 0)으로 설정하면 sigma 값에 의해 자동으로 설정
# sigmaX: X방향 sigma, 0이면 ksize에 따라 자동 계산
# sigmaY: Y방향 sigma, 0이면 sigmaX와 동일
result1 = cv2.GaussianBlur(src=img, ksize=(0, 0), sigmaX=1)
result3 = cv2.GaussianBlur(img, (0, 0), 3)

# 평균필터와 비교
resultM = cv2.blur(img, (5, 5))
# ksize 직접 설정
resultG = cv2.GaussianBlur(img, (5, 5), 0)

cv2.imshow('Original', img)
cv2.imshow('Sigma 1', result1)
cv2.imshow('Sigma 3', result3)
cv2.imshow('Mean Filter', resultM)
cv2.imshow('Gaussian Filter', resultG)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./data/Lenna.png', cv2.IMREAD_GRAYSCALE)
# 직접 만들어서 적용
# 상하좌우 + 가운데 강조
kernel1 = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
result1 = cv2.filter2D(img, -1, kernel1)
# 상하좌우 + 가운데 강조 + 대각선
kernel2 = np.array([[-1, -1, -1],
                    [-1, 9, -1],
                    [-1, -1, -1]])
result2 = cv2.filter2D(img, -1, kernel2)

cv2.imshow('Original', img)
cv2.imshow('Sharpening 1', result1)
cv2.imshow('Sharpening 2', result2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./data/Lenna.png', cv2.IMREAD_GRAYSCALE)

blur = cv2.GaussianBlur(img, (0, 0), 2)
# 원 영상에서 블러링(부드러운 성분) 빼서 선명하게 만들기
result = np.clip(2 * img - blur, 0, 255).astype(np.uint8)

cv2.imshow('Original', img)
cv2.imshow('Unsharpening', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('./data/noise1.jpg', cv2.IMREAD_GRAYSCALE)
# src: 입력영상
# d: 이웃 픽셀 거리, -1인 경우 자동 계산
# sigmaColor: 색 공간에서 가우시안 필터의 표준편차, 값이 클수록 더 많은 주변 픽셀을 고려하여 블러링
# sigmaSpace: 좌표 공간에서 가우시안 필터의 표준편차, 값이 작으면 픽셀값 차이가 크면 블러링 안됨, 값이 크면 픽셀값 차이가 조금 크더라도 블러링 적용됨, 즉 어느 정도 밝기 차를 갖는 경계선을 보존할 것인지 조정할 수 있음
resultB = cv2.bilateralFilter(src=img, d=9, sigmaColor=75, sigmaSpace=75)
# resultB = cv2.bilateralFilter(img, 3, 75, 75)
# resultB = cv2.bilateralFilter(img, -1, 75, 75)
# resultB = cv2.bilateralFilter(img, -1, 10, 10)
# 평균필터, 가우시안 필터와 비교
resultM = cv2.blur(img, (5, 5))
resultG = cv2.GaussianBlur(img, (5, 5), 0)

cv2.imshow('Original', img)
cv2.imshow('Bilateral Filter', resultB)
cv2.imshow('Mean Filter', resultM)
cv2.imshow('Gaussian Filter', resultG)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
